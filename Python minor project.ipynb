```python
sns.set(style='whitegrid')
ed_percentage = file.groupby(['Education_Level', 'Attrition_Flag']).size().unstack(fill_value=0)
ed_attr_percentage = ed_attr_percentage.div(ed_attr_percentage.sum(axis=1), axis=0) * 100
plt.figure(figsize=(12, 8))
ed_attr_percentage.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Percentage Segmented Bar Graph between Education_Level and Attrition_Flag')
plt.xlabel('Education_Level')
plt.ylabel('Percentage')
plt.legend(title='Attrition_Flag', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
income_attrition_percentage = file.groupby(['Income_Category', 'Attrition_Flag']).size().unstack(fill_value=0)
income_attrition_percentage = income_attrition_percentage.div(income_attrition_percentage.sum(axis=1), axis=0) * 100
sns.set(style='whitegrid')
plt.figure(figsize=(12, 8))
income_attrition_percentage.plot(kind='bar', stacked=True, colormap='Set2')
plt.title('Percentage Segmented Bar Graph between Income_Category and Attrition_Flag')
plt.xlabel('Income_Category')
plt.ylabel('Percentage')
plt.legend(title='Attrition_Flag', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
